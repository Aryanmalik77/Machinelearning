```python
from sklearn.datasets import load_diabetes
import numpy as np
from sklearn.metrics import r2_score
```

```python
X, y = load_diabetes (return_X_y= True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,y , test_size =0.2, random_state = 2)
```

```python
# using sklearn
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.001)
```

```python
reg.fit(X_train, Y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
r2_score(Y_test, y_pred)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
class MeraRidgeReg:
    def __init__(self, learning_rate, epochs , alpha):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.alpha = alpha
        self.m =None
        self.b = None
    def fit(self,X_train, Y_train):
        self.coef_ = np.ones(X_train.shape[1])
        self.intercept_= 0
        theta = np.insert(self.coef_, 0, self.intercept_)
        X_train = np.insert(X_train, 0, 1, axis=1)
        for i in range (self.epochs):
            theta_dr = np.dot(X_train.T ,X_train).dot(theta ) -np.dot(X_train.T, Y_train) + self.alpha*theta 
            theta = theta - self.learning_rate*theta_dr
        self.coef_ = theta[1:]
        self.intercept_ = theta[0]
    def predict(self,X_test):
        
        return np.dot(X_test,self.coef_) + self.intercept_
            
            


        

    
```

```python
reg = MeraRidgeReg(epochs=500,alpha=0.001,learning_rate=0.005)
```

```python
reg.fit(X_train,Y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(Y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```
