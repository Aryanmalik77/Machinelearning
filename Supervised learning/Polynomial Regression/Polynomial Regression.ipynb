```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 6*np.random.rand(200,1)
y = 0.8*x**2 +0.9*x + 2 + np.random.randn(200,1) 
```

```python
plt.plot(x,y,"b.")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size =0.25, random_state = 43)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, y_pred)
```

```python
plt.plot(X_train,Y_train,"b.")
plt.plot(X_train,lr.predict(X_train), color ="r")
```

```python
# by using the polynomial linear Regression
```

```python
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
```

```python
pl = PolynomialFeatures(degree =2)
```

```python
X_train_trans = pl.fit_transform(X_train)
X_test_trans = pl.transform(X_test)
```

```python
#include baise parameter

```

```python
lr = LinearRegression()
lr.fit(X_train_trans,Y_train)
```

```python
y_pred = lr.predict(X_test_trans)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, y_pred)
```

```python
X_new=np.linspace(-7, 7, 200).reshape(200, 1)
X_new_poly = pl.transform(X_new)
y_new = lr.predict(X_new_poly)
```

```python
plt.plot(X_train,Y_train,"b.") 
plt.plot(X_new,y_new,color = "red", linewidth = "2" )
```
